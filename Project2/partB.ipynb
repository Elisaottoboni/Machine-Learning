```python
from sklearn.datasets import load_breast_cancer
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
from sklearn.model_selection import train_test_split, KFold
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import roc_curve, auc
import numpy as np
import matplotlib.pyplot as plt
from FFNN_class import *

#%matplotlib inline

# Load dataset
wisconsin = load_breast_cancer()
X = wisconsin.data
target = wisconsin.target
target = target.reshape(target.shape[0], 1)

# Normalize the data
scaler = MinMaxScaler()
X = scaler.fit_transform(X)

# Set up k-fold cross-validation
kf = KFold(n_splits=10, shuffle=True, random_state=2024)

# Initialize arrays to hold the results for ROC
tpr_list = []  # True Positive Rates
fpr_list = []  # False Positive Rates
thresholds_list = []  # Thresholds

for train_index, test_index in kf.split(X):
    X_train, X_val = X[train_index], X[test_index]
    t_train, t_val = target[train_index], target[test_index]

    # Define the FFNN model for logistic regression
    input_nodes = X_train.shape[1]
    output_nodes = 1
    logistic_regression = FFNN((input_nodes, output_nodes), output_func=sigmoid, cost_func=CostLogReg, seed=2023)
    logistic_regression.reset_weights()

    # Train the model
    scheduler = Adam(eta=1e-3, rho=0.9, rho2=0.999)
    logistic_regression.fit(X_train, t_train, scheduler, epochs=1000, X_val=X_val, t_val=t_val)

    # Predictions on the validation set
    predictions = logistic_regression.predict(X_val)
    predictions_prob = predictions.flatten()

    # Compute ROC curve
    fpr, tpr, thresholds = roc_curve(t_val, predictions_prob)
    tpr_list.append(tpr)
    fpr_list.append(fpr)
    thresholds_list.append(thresholds)

# Average the ROC metrics
mean_tpr = np.mean(tpr_list, axis=0)
mean_fpr = np.mean(fpr_list, axis=0)

# Plot the ROC curve
plt.figure()
plt.plot(mean_fpr, mean_tpr, color='b', label='Mean ROC curve (area = %0.2f)' % auc(mean_fpr, mean_tpr))
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import roc_curve, auc
from scipy import interpolate
from FFNN_class import *

%matplotlib inline

# Load dataset
wisconsin = load_breast_cancer()
X = wisconsin.data
target = wisconsin.target
target = target.reshape(target.shape[0], 1)

# Normalize the data
scaler = MinMaxScaler()
X = scaler.fit_transform(X)

# Set parameters
n_splits = 10
random_state = 2023
test_size = 0.2

# Initialize arrays to hold the results for ROC
tpr_list = []  # True Positive Rates
fpr_list = []  # False Positive Rates

# Run multiple splits
for _ in range(n_splits):
    X_train, X_val, t_train, t_val = train_test_split(X, target, test_size=test_size, random_state=random_state)

    # Define and train logistic regression model
    input_nodes = X_train.shape[1]
    output_nodes = 1
    logistic_regression = FFNN((input_nodes, output_nodes), output_func=sigmoid, cost_func=CostLogReg, seed=random_state)

    # Reset weights
    logistic_regression.reset_weights()

    # Train the model
    scheduler = Adam(eta=1e-3, rho=0.9, rho2=0.999)
    logistic_regression.fit(X_train, t_train, scheduler, epochs=1000, X_val=X_val, t_val=t_val)

    # Predictions on the validation set
    predictions_prob = logistic_regression.predict(X_val).flatten()

    # Compute ROC curve
    fpr, tpr, _ = roc_curve(t_val, predictions_prob)
    fpr_list.append(fpr)
    tpr_list.append(tpr)

# Interpolating the TPR at common FPR values
mean_fpr = np.linspace(0, 1, 100)
mean_tpr = np.zeros_like(mean_fpr)

for i in range(n_splits):
    tpr_interp = interpolate.interp1d(fpr_list[i], tpr_list[i], bounds_error=False, fill_value="extrapolate")
    mean_tpr += tpr_interp(mean_fpr)

mean_tpr /= n_splits

# Plot the average ROC curve for logistic regression
plt.figure()
plt.plot(mean_fpr, mean_tpr, color='b', label='Mean ROC curve (Logistic Regression)')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Logistic Regression')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
# Neural Network Evaluation with different activation functions
activation_functions = {
    "ReLU": (RELU, CostCrossEntropy),
    "LRELU": (LRELU, CostCrossEntropy),
    "Sigmoid": (sigmoid, CostCrossEntropy),
}

for act_name, (hidden_func, cost_func) in activation_functions.items():
    tpr_list = []
    fpr_list = []

    for _ in range(n_splits):
        X_train, X_val, t_train, t_val = train_test_split(X, target, test_size=test_size, random_state=random_state)

        # Define and train neural network model
        hidden_nodes1 = 100
        hidden_nodes2 = 30
        dims = (input_nodes, hidden_nodes1, hidden_nodes2, output_nodes)
        neural_network = FFNN(dims, hidden_func=hidden_func, output_func=sigmoid, cost_func=cost_func, seed=random_state)

        # Reset weights
        neural_network.reset_weights()

        # Train the model
        scheduler = Adam(eta=1e-4, rho=0.9, rho2=0.999)
        neural_network.fit(X_train, t_train, scheduler, epochs=1000, X_val=X_val, t_val=t_val)

        # Predictions on the validation set
        predictions_prob = neural_network.predict(X_val).flatten()

        # Compute ROC curve
        fpr, tpr, _ = roc_curve(t_val, predictions_prob)
        fpr_list.append(fpr)
        tpr_list.append(tpr)

    # Interpolating the TPR at common FPR values for the neural network
    mean_tpr = np.zeros_like(mean_fpr)

    for i in range(n_splits):
        tpr_interp = interpolate.interp1d(fpr_list[i], tpr_list[i], bounds_error=False, fill_value="extrapolate")
        mean_tpr += tpr_interp(mean_fpr)

    mean_tpr /= n_splits

    # Plot the average ROC curve for the neural network
    plt.plot(mean_fpr, mean_tpr, label=f'Mean ROC curve ({act_name})')

# Finalizing the plot
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves - Neural Networks with Different Activation Functions')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.legend(loc='lower right')
plt.grid()
plt.show()
```
