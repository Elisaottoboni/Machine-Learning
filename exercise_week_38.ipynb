# Analytical Exercise

## Context

Consider a dataset $\mathcal{L}$ consisting of the data $\mathbf{X}_{\mathcal{L}} = \{(y_j, \mathbf{x}_j), j = 0 \dots n-1\}$.

We assume that the true data is generated from a noisy model

$$
\mathbf{y} = f(x) + \epsilon
$$

Here $\epsilon$ is normally distributed with mean zero and standard deviation $\sigma^2$.

In our derivation of the ordinary least squares method, we defined an approximation to the function $f$ in terms of the parameters $\boldsymbol{\beta}$ and the design matrix $\mathbf{X}$ which embody our model, that is $\mathbf{\tilde{y}} = \mathbf{X}\boldsymbol{\beta}$.

The parameters $\boldsymbol{\beta}$ are in turn found by optimizing the mean squared error via the so-called cost function

$$
C(\mathbf{X}, \boldsymbol{\beta}) = \frac{1}{n} \sum_{i=0}^{n-1} (y_i - \tilde{y}_i)^2 = \mathbb{E} \left[ (\mathbf{y} - \mathbf{\tilde{y})}^2 \right]
$$

Here the expected value $\mathbb{E}$ is the sample value.

## Request

Our task is to show that 

$$
\mathbb{E} \left[ (\mathbf{y} - \tilde{\mathbf{y}})^2 \right] = \text{Bias}[\tilde{\mathbf{y}}] + \text{Var}[\tilde{\mathbf{y}}] + \sigma^2
$$

with 

$$
\text{Bias}[\tilde{\mathbf{y}}] = \mathbb{E} \left[ (\mathbf{y} - \mathbb{E}[\tilde{\mathbf{y}}])^2 \right]
$$

and

$$
\text{var}[\tilde{\mathbf{y}}] = \mathbb{E} \left[ (\tilde{\mathbf{y}} - \mathbb{E}[\tilde{\mathbf{y}}])^2 \right] = \frac{1}{n} \sum_i (\tilde{y}_i - \mathbb{E}[\tilde{\mathbf{y}}])^2
$$


## Solution

First, it is worth noting that by rewriting the cost function as an expectation (rather than a sample mean), we are generalizing it. That is, we are considering a broader probabilistic framework, not just a finite dataset, but a distribution over all possible outcomes (the true distribution of $\mathbf{y}$). So we obtain the following equation:

$$
C(\mathbf{X}, \boldsymbol{\beta}) = \frac{1}{n} \sum_{i=0}^{n-1} (y_i - \tilde{y}_i)^2 = \mathbb{E} \left[ (\mathbf{y} - \mathbf{\tilde{y})}^2 \right]
$$

This notation also allows us to break the total error into bias, variance, and noise components for the true model (which we can't directly observe) rather than just the error on a single dataset.

In order to show that:

$$
\mathbb{E} \left[ (\mathbf{y} - \tilde{\mathbf{y}})^2 \right] = \text{Bias}[\tilde{\mathbf{y}}] + \text{Var}[\tilde{\mathbf{y}}] + \sigma^2
$$

we start by writing the cost function as an expactation:

$$
C(\mathbf{X}, \boldsymbol{\beta}) = \mathbb{E} \left[ (\mathbf{y} - \mathbf{\tilde{y}})^2 \right]
$$

We expand $ (\mathbf{y} - \tilde{\mathbf{y}})^2 $ using the fact that $ \mathbf{y} = f(\mathbf{x}) + \epsilon $:

$$
 \mathbb{E} \left[ (\mathbf{y} - \tilde{\mathbf{y}})^2 \right] =  \mathbb{E} \left[ \left( f(\mathbf{x}) + \epsilon - \tilde{\mathbf{y}} \right)^2 \right]
$$

Then, we add and subtract $\mathbb{E}[\mathbf{\tilde{y}}]$ to the above equation:

$$
\mathbb{E} \left[ (\mathbf{\mathbf{y} - \mathbf{\tilde{y}}})^2 \right] = \mathbb{E} \left[ \left( f(\mathbf{x}) + \epsilon - \mathbf{\tilde{y}} + \mathbb{E}[\mathbf{\tilde{y}}] - \mathbb{E}[\mathbf{\tilde{y}}] \right)^2 \right]
$$

Finally, by applying the linearity of expectation, we can rewrite the above formula as follows:

$$
\mathbb{E} \left[ (\mathbf{y} - \mathbf{\tilde{y}})^2 \right] = \mathbb{E} \left[ \left( f(\mathbf{x}) - \mathbb{E}[\mathbf{\tilde{y}}] \right)^2 \right] + \mathbb{E} \left[ \left( \mathbf{\tilde{y}} - \mathbb{E}[\mathbf{\tilde{y}}] \right)^2 \right] + \sigma^2.
$$

Now we can rewrite the terms as:

- $\text{Bias}[\tilde{\mathbf{y}}] = \mathbb{E} \left[ (f(\mathbf{x}) - \mathbb{E}[\tilde{\mathbf{y}}])^2 \right]$. It measures how much the average prediction ($\mathbb{E}[\tilde{\mathbf{y}}]$) differs from the true function value ($f(\mathbf{x})$).

- $\text{Var}[\tilde{\mathbf{y}}] = \mathbb{E} \left[ (\tilde{\mathbf{y}} - \mathbb{E}[\tilde{\mathbf{y}}])^2 \right]$.  It measures how much the predictions $\tilde{\mathbf{y}}$ vary around their expected value $\mathbb{E}[\tilde{\mathbf{y}}]$.

- **Noise term**: $\sigma^2$. This is the irreducible noise in the data, that we assume follows a normal distribution.

Thus, the final expression is:

$$
\mathbb{E} \left[ (\mathbf{y} - \tilde{\mathbf{y}})^2 \right] =  \text{Bias}[\tilde{\mathbf{y}}] + \text{Var}[\tilde{\mathbf{y}}] + \sigma^2.
$$

## Request

Perform then a bias-variance analysis of a simple one-dimensional (or other models of your choice) function by studying the MSE value as function of the complexity of your model. Use ordinary least squares only.

Discuss the bias and variance trade-off as function of your model complexity (the degree of the polynomial) and the number of data points, and possibly also your training and test data using the \textbf{bootstrap} resampling method. ## Solution

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.utils import resample

# Define the true function (quadratic) and dataset sizes
def generate_data(n):
    x = np.linspace(-3, 3, n).reshape(-1, 1)  # Input feature x
    y = 2.0 + 5*x**2 + 0.1*np.random.randn(n, 1)  # Quadratic function with noise
    return x, y

# Parameters for the analysis
n_boostraps = 500  # Number of bootstrap resamples
maxdegree = 14  # Maximum polynomial degree
dataset_sizes = [30, 50, 100, 200]  # Different dataset sizes

# Loop over different dataset sizes
for n in dataset_sizes:
    print(f'\nAnalyzing for dataset size: {n}')
    x, y = generate_data(n)
    
    error = np.zeros(maxdegree)
    bias = np.zeros(maxdegree)
    variance = np.zeros(maxdegree)
    polydegree = np.zeros(maxdegree)
    
    # Split the dataset into training and test sets with shuffle
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True)
    
    # Loop over different polynomial degrees
    for degree in range(maxdegree):
        # Include StandardScaler in the pipeline
        model = make_pipeline(
            PolynomialFeatures(degree=degree),
            StandardScaler(),
            LinearRegression(fit_intercept=True)
        )
        
        y_pred = np.empty((y_test.shape[0], n_boostraps))  # Store predictions for each bootstrap iteration
        
        # Perform bootstrap resampling
        for i in range(n_boostraps):
            x_resampled, y_resampled = resample(x_train, y_train)  # Resample the training data
            model.fit(x_resampled, y_resampled)  # Fit the model on resampled data
            y_pred[:, i] = model.predict(x_test).ravel()  # Predict on the test set
        
        # ERRORE ENORME NEL CALCOLO DELL'ERRORE, DEL BIAS E DELLA VARIANZA.
        # Calculate error, bias, and variance
        polydegree[degree] = degree
        error[degree] = np.mean((y_test.ravel() - y_pred.mean(axis=1))**2)
        bias[degree] = np.mean((y_test.ravel() - y_pred.mean(axis=1))**2)
        variance[degree] = np.mean(y_pred.var(axis=1))
        
        # Print results for each degree
        print(f'Polynomial degree: {degree}')
        print(f'Error: {error[degree]}')
        print(f'Bias^2: {bias[degree]}')
        print(f'Variance: {variance[degree]}')
        print(f'{error[degree]} ≈ {bias[degree]} + {variance[degree]} = {bias[degree] + variance[degree]}')
    
    # Plot the bias-variance trade-off for the current dataset size
    plt.figure(figsize=(10, 6))
    plt.plot(polydegree, error, label='Error', marker='o')
    plt.plot(polydegree, bias, label='Bias^2', marker='s')
    plt.plot(polydegree, variance, label='Variance', marker='^')
    plt.title(f'Bias-Variance Trade-off for Dataset Size: {n}')
    plt.xlabel('Polynomial Degree')
    plt.ylabel('Error')
    plt.legend()
    plt.grid(True)
    plt.show()
```

**Considerations**

When it comes to the bias-variance trade off, we expect models characterized by lower polynomial degrees to result in higher bias, but lower variance. This is because the model makes strong assumptions (like linearity), causing an increase in the overall bias level.  On the other hand, models that present a higher polynomial degree tend to cause overfitting, meaning they fit the training data almost to perfection. For this reason, we would expect a lower bias, but higher variance in the final results.
